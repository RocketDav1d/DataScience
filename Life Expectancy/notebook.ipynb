```python
def draw_scatterplot(year):
    life_expectancy = df1[year]
    fertility_rate = df3[year]

    new_df = pd.DataFrame({
    'Life Expectancy' + str(year): life_expectancy,
    'Fertility Rate' + str(year): fertility_rate
    })

    plt.scatter(new_df['Life Expectancy' + str(year)], new_df['Fertility Rate' + str(year)])
    plt.title('Fertility Rate vs Life Expectancy in ' + str(year))
    plt.xlabel('Life Expectancy')
    plt.ylabel('Fertility Rate')
    plt.xlim(20, 90)
    plt.ylim(0, 10)
    plot = plt.gcf()
    return plot
    
```

```python
def save_plots(year1, year2, filename):
    for year in range(year1, year2):
        plot = draw_scatterplot(str(year))
        plot.savefig(filename + str(year) + '.png', dpi=300)
        plt.close(plot)
```

```python
import math

def save_plots(year1, year2, filename):
    num_years = year2 - year1
    num_cols = 3  # Number of columns for subplots
    num_rows = math.ceil(num_years / num_cols)  # Calculate number of rows

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 5))

    for idx, year in enumerate(range(year1, year2)):
        row = idx // num_cols
        col = idx % num_cols
        ax = axes[row, col]
        
        life_expectancy = df1[str(year)]
        fertility_rate = df3[str(year)]

        new_df = pd.DataFrame({
            'Life Expectancy': life_expectancy,
            'Fertility Rate': fertility_rate
        })

        ax.scatter(new_df['Life Expectancy'], new_df['Fertility Rate'])
        ax.set_title('Fertility Rate vs Life Expectancy in ' + str(year))
        ax.set_xlabel('Life Expectancy')
        ax.set_ylabel('Fertility Rate')
        ax.set_xlim(20, 90)
        ax.set_ylim(0, 10)

    # Adjust the layout to prevent overlapping
    plt.tight_layout()
    # Save the figure with all subplots to a single file
    fig.savefig(filename + '.png', dpi=300)
    plt.close(fig)
```

```python
save_plots(1960, 2010, 'scatterplot')
```

```python
import os
def save_scatterplot(year, folder):
    if not os.path.exists(folder):
        os.makedirs(folder)

    life_expectancy = df1[str(year)]
    fertility_rate = df3[str(year)]

    new_df = pd.DataFrame({
        'Life Expectancy': life_expectancy,
        'Fertility Rate': fertility_rate
    })

    plt.scatter(new_df['Life Expectancy'], new_df['Fertility Rate'])
    plt.title('Fertility Rate vs Life Expectancy in ' + str(year))
    plt.xlabel('Life Expectancy')
    plt.ylabel('Fertility Rate')
    plt.xlim(20, 90)
    plt.ylim(0, 10)
    plt.savefig(f"{folder}/scatterplot_{year}.png", dpi=300)
    plt.close()
```

```python
def save_all_scatterplots(year1, year2, folder):
    for year in range(year1, year2 + 1):
        save_scatterplot(year, folder)
```

```python
save_all_scatterplots(1960, 2010, "scatterplots")
```

```python
import imageio.v2 as imageio

def create_animation(year1, year2, input_folder, output_filename):
    images = [
        imageio.imread(f"{input_folder}/scatterplot_{year}.png")
        for year in range(year1, year2 + 1)
    ]
    
    imageio.mimsave(output_filename, images, fps=20)

create_animation(1960, 2010, "scatterplots", "animation.gif")
```

```python
def convert_value(value_str):
    multipliers = {
        'k': 1_000,
        'M': 1_000_000,
        'B': 1_000_000_000,
    }

    # Extract the numeric value and the multiplier letter, if any
    value_num, multiplier_letter = float(value_str[:-1]), value_str[-1]

    # Check if the last character is a multiplier letter
    if multiplier_letter in multipliers:
        value_num *= multipliers[multiplier_letter]
    else:
        # If the last character is not a multiplier letter, it is part of the number
        value_num = float(value_str)

    return value_num
```

```python
def pop_convert(x):
    if x.endswith('k'):
        return int(float(x[:-1]) * 1_000)
    elif x.endswith('M'):
        return int(float(x[:-1]) * 1_000_000)
    elif x.endswith('B'):
        return int(float(x[:-1]) * 1_000_000_000)
    else:
        return int(x)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_scatter_by_population(df, year, output_file, normalize_min=10, normalize_max=200):
    year = str(year)  # Make sure the year is a string for column indexing
    x = df['country']  # Replace 'country' with your x-axis data column name, if different
    y = df[year]
    continents = df['continents']

    # Normalize the population values to the desired range for bubble size
    population_range = y.max() - y.min()
    normalized_population = ((y - y.min()) / population_range) * (normalize_max - normalize_min) + normalize_min

    # Create a dictionary that maps each continent to a unique color
    continent_color = {}
    unique_continents = df['continents'].unique()
    colors = np.random.choice(list(plt.cm.colors.CSS4_COLORS.keys()), len(unique_continents), replace=False)
    for i, continent in enumerate(unique_continents):
        continent_color[continent] = colors[i]

    plt.figure(figsize=(12, 6))
    for i, (x_val, y_val, continent) in enumerate(zip(x, y, continents)):
        plt.scatter(x_val, y_val, s=normalized_population[i], alpha=0.5, c=continent_color[continent], label=continent if i == 0 else "")
    plt.xlabel('Country')
    plt.ylabel('Population')
    plt.xticks(rotation=90)
    plt.title(f'Scatter plot of population for the year {year}')
    plt.legend()
    plt.tight_layout()
    plt.savefig(output_file)
    plt.show()

# Usage example
plot_scatter_by_population(converted_population_total_result_df, 2000, 'scatterplot_2000.png')
```
