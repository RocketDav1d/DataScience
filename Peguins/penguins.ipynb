```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='species', y='bill_length_mm', data=df)
plt.title('Beak Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Beak Length (mm)')
plt.show()

sns.boxplot(x='species', y='bill_depth_mm', data=df)
plt.title('Beak Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Beak Width (mm)')
plt.show()
```

```python
sns.violinplot(x='species', y='bill_length_mm', data=df)
plt.title('Beak Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Beak Length (mm)')
plt.show()

sns.violinplot(x='species', y='bill_depth_mm', data=df)
plt.title('Beak Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Beak Width (mm)')
plt.show()
```

Chinstrap Species have the second largest maximum beak width and length. On average the animals of this species probably have the overall biggest beaks as the distributuion for both length and width is quite even.

Gento's have the longest beak, but only very few animals, have such a long beak. They also have the smallest beak in terms of width, with a  pretty evely distribution

Generally the beak width for all species is quite evenly distributed, especially compared to the beak length. There only a tiny few have the maximum beak length and the majority of beak lengths is closer to the minimum

```python
avg_beak_dims = df.groupby('species')[['bill_length_mm', 'bill_depth_mm']].mean()
```

```python
print(avg_beak_dims)
```

Based on the average beak dimensions for each species, we can derive the following key insights:

Adelie penguins have the shortest beaks, with an average length of 38.79 mm and an average depth of 18.35 mm. This indicates that they have relatively short and wide beaks compared to the other species.

Chinstrap penguins have beaks that are longer than Adelie penguins, with an average length of 48.83 mm. However, their beak depth is almost the same as Adelie penguins, averaging 18.42 mm. This suggests that Chinstrap penguins have beaks that are relatively long and moderately wide.

Gentoo penguins have the longest and slimmest beaks among the three species, with an average length of 47.50 mm and the smallest average depth of 14.98 mm. This implies that Gentoo penguins have long and slim beaks.

In summary, Gentoo penguins have long and slim beaks, while Adelie penguins have rather short and wide beaks. Chinstrap penguins have beaks that are comparatively longer than Adelie penguins but with a similar width.
